## choose features and construct a new dataframe

```python
df1 = pd.DataFrame()
df1['open']=df['open']
df1['volume']=df['volume']
df1['quote_volume']=df['quote_volume']
df1['trades']=df['trades']
df1['taker_base_volue']=df['taker_base_volue']
df1['taker_quote_volume']=df['taker_quote_volume']
df1[:10]
```

## normalize and split training set and test set

```python
preprocessing = MinMaxScaler()
data = preprocessing.fit_transform(df1.values)
```

```python
train_size = int(len(data)*0.8)
training_set = data[:train_size]
test_set = data[train_size:]
```

```python
predict_min = 30
```

### we want to predict the price after 30 min

```python
#each data record 1 min data, we need to reshape they as a block which has 30 data. 
#And use block to predict next price after 30 min
X_train = []
Y_train = []
for i in range(predict_min, training_set.shape[0]):
    X_train.append(training_set[i-predict_min:i,:])
    Y_train.append(training_set[i,:])   
X_train, Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_test = []
Y_test = []
for i in range(predict_min, test_set.shape[0]):
    X_test.append(test_set[i-predict_min:i,:])
    Y_test.append(test_set[i,:])
X_test, Y_test = np.array(X_test), np.array(Y_test)
```

# construct model

```python
model = Sequential()
```

```python
model.add(LSTM(50, input_shape = (X_train.shape[1],X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units =50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units =50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dense(6))
```

```python
model.summary()
```

```python
model.compile(loss="mse", optimizer="adam")
```

```python
model.fit(X_train,Y_train, epochs=10, batch_size= 100)
```

### use test data to predict and construct a new dataframe to compare

```python
predict_price = model.predict(X_test)
predict_price = preprocessing.inverse_transform(predict_price)
df2 = pd.DataFrame(predict_price)
real_price = preprocessing.inverse_transform(Y_test)
df3 = pd.DataFrame(real_price)
```

```python
result = pd.DataFrame()
result['time'] = df['time'][train_size:][predict_min:len(test_set)]
result['real_price'] = np.array(df2.iloc[:,0])
result['predict_price']= np.array(df3.iloc[:,0])
result.head()
```

```python
plt.figure(figsize=(20,10))
ax=plt.gca()
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
plt.xticks(rotation=45)
plt.plot(result["time"], result['predict_price'],color='red',label='predict_price',linewidth=0.5)
plt.plot(result["time"], result['real_price'],color='blue',label='real_price',linewidth=0.5)
plt.legend(prop={'size': 20})
plt.show()
```

# interactive plot the predict price and real price

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
result=result.reset_index()
```

```python
def plotfigure(x,y):
    plt.figure(figsize=(20,10))
    ax=plt.gca()
    xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
    ax.xaxis.set_major_formatter(xfmt)
    plt.xticks(rotation=45)
    plt.plot(result["time"][x:y], result['predict_price'][x:y],color='red',label='predict_price')
    plt.plot(result["time"][x:y], result['real_price'][x:y],color='blue',label='real_price')
    plt.legend(prop={'size': 20})
    plt.title('From '+ str(result["time"][x])+ ' to '+ str(result["time"][y]),fontsize=35)
    plt.show()
```

```python
interactive_plot = interactive(plotfigure, x=(0,len(result)-1), y=(0,len(result)-1))
interactive_plot
```
