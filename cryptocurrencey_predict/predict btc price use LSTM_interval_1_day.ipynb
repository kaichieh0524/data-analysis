## choose features and construct a new dataframe

```python
df1 = pd.DataFrame()
df1['open']=df['open']
df1['volume']=df['volume']
df1['quote_volume']=df['quote_volume']
df1['trades']=df['trades']
df1['taker_base_volue']=df['taker_base_volue']
df1['taker_quote_volume']=df['taker_quote_volume']
df1.head()
```

## normalize and split training set and test set

```python
preprocessing = MinMaxScaler()
data = preprocessing.fit_transform(df1.values)
```

```python
train_size = int(len(data)*0.8)
training_set = data[:train_size]
test_set = data[train_size:]
```

### we want to predict the price after 7 days

```python
predict_day = 7
X_train = training_set[0:len(training_set) - predict_day]
Y_train = training_set[predict_day:len(training_set)]
X_test = test_set[0:len(test_set) - predict_day]
Y_test = test_set[predict_day:len(test_set)]
```

### reshape in order to fit LSTM model

```python
X_train = np.reshape(X_train, (len(X_train), 1 , X_train.shape[1]))
X_test = np.reshape(X_test, (len(X_test), 1 , X_test.shape[1]))
```

# construct model

```python
model = Sequential()
```

```python
model.add(LSTM(50, input_shape = (X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units =50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units =50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dense(6))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(X_train,Y_train, epochs=20, batch_size= 8)
```

### use test data to predict and construct a new dataframe to compare

```python
predict_price = model.predict(X_test)
predict_price = preprocessing.inverse_transform(predict_price)
df2 = pd.DataFrame(predict_price)
real_price = preprocessing.inverse_transform(Y_test)
df3 = pd.DataFrame(real_price)
```

```python
result = pd.DataFrame()
result['time'] = df['time'][train_size:][predict_day:len(test_set)]
result['real_price'] = np.array(df2.iloc[:,0])
result['predict_price']= np.array(df3.iloc[:,0])
result.head()
```
